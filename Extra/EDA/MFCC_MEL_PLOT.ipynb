```python
import matplotlib.pyplot as plt
import librosa
import librosa.display
import numpy as np


```

```python
# Load the audio file
filename = '/home/tnguyen/Downloads/sensor1.wav'
y, sr = librosa.load(filename)

```

```python
# Generate Mel spectrogram
S = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=2048, hop_length=512, n_mels=128)
S_DB = librosa.power_to_db(S, ref=np.max)

# Generate MFCC
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)

```

```python
# Create figure
fig, ax = plt.subplots(2, 1, figsize=(10, 10))

# Display Mel spectrogram
img1 = librosa.display.specshow(S_DB, sr=sr, hop_length=512, x_axis='time', y_axis='mel', ax=ax[0])
ax[0].set_title('Mel-Spectrogram Sensor 1')
fig.colorbar(img1, ax=ax[0], format="%+2.f dB")

# Display MFCC
img2 = librosa.display.specshow(mfccs, sr=sr, x_axis='time', ax=ax[1])
ax[1].set_title('MFCC Sensor 1')
fig.colorbar(img2, ax=ax[1])

plt.tight_layout()
plt.savefig('mel_mfcc.png')
plt.show()
```
