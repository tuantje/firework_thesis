```python
#/home/tnguyen/Downloads/
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
from tqdm import tqdm

# File names
file_names = ["sensor1.wav", "sensor5.wav", "sensor9.wav"]
```

```python
# Initialize the plots
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Loop over the files
for i in tqdm(range(3)):
    # Load the audio file
    y, sr = librosa.load(f"/home/tnguyen/Downloads/{file_names[i]}")
    
    # Compute MFCC features
    mfccs = librosa.feature.mfcc(y=y, sr=sr)
    
    # Display the MFCCs
    librosa.display.specshow(mfccs, sr=sr, x_axis='time', ax=axs[i])
    
    # Set the title
    axs[i].set_title(f"MFCC Sensor {file_names[i][6]}", fontsize=16)
    
    # Set the colorbar
    fig.colorbar(axs[i].collections[0], ax=axs[i])
    
plt.tight_layout()
plt.show()
```
