```python
import matplotlib.pyplot as plt

# Intensity percentages
intensities = [100, 54.45, 56.69, 39.81, 18.24, 28.07, 18.81, 12.64, 16.74]

# Corresponding sensors
sensors = ['Sensor 1','Sensor 2', 'Sensor 3', 'Sensor 4', 'Sensor 5', 'Sensor 6', 'Sensor 7', 'Sensor 8', 'Sensor 9']

# Create the plot
plt.figure(figsize=(10, 6))
plt.bar(sensors, intensities, color='firebrick')
plt.xlabel('Sensors')
plt.ylabel('Intensity Relative to Sensor 1 (%)')
plt.title('Relative Intensities of Sound Detected by Each Sensor Compared to Sensor 1')
plt.ylim(0, 100)  # This makes sure the y-axis represents the full percentage range (0 to 100%)
plt.show()
```
