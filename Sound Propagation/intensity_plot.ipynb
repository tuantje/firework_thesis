```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="darkgrid")
sns.set_color_codes("pastel")

# List of sensors
sensors = ['s1', 's2', 's3', 's4', 's5', 's6', 's7', 's8', 's9']

# Intensity values corresponding to each sensor
intensities = [100, 54.45, 56.69, 39.81, 18.24, 28.07, 18.81, 12.64, 16.74]

# Create a barplot
plt.figure(figsize=(10,6))
sns.barplot(x=sensors, y=intensities, palette="Blues_d")

# Add title and labels
plt.title('Intensity at each sensor compared to sensor 1', fontsize=18)
plt.xlabel('Sensor', fontsize=15)
plt.ylabel('Intensity (%)', fontsize=15)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set(style="darkgrid")

# List of sensors
sensors = ['s1', 's2', 's3', 's4', 's5', 's6', 's7', 's8', 's9']

# Intensity values corresponding to each sensor
intensities = [100, 54.45, 56.69, 39.81, 18.24, 28.07, 18.81, 12.64, 16.74]

# Create a color array using a colormap
colors = plt.cm.viridis(np.linspace(0, 1, len(sensors)))

# Create a bar plot
plt.figure(figsize=(10,6))
plt.bar(sensors, intensities, color=colors)

# Add title and labels
plt.title('Intensity at each sensor compared to sensor 1', fontsize=18)
plt.xlabel('Sensor', fontsize=15)
plt.ylabel('Intensity (%)', fontsize=15)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set(style="darkgrid")

# List of sensors
sensors = ['s1', 's2', 's3', 's4', 's5', 's6', 's7', 's8', 's9']

# Intensity values corresponding to each sensor
intensities = [100, 54.45, 56.69, 39.81, 18.24, 28.07, 18.81, 12.64, 16.74]

# Create a color array using a colormap
colors = plt.cm.viridis(np.linspace(0, 1, len(sensors)))

# Create a bar plot
plt.figure(figsize=(10,6))
plt.bar(sensors, intensities, color=colors)

# Add the line plot on top of the bars
sns.lineplot(x=sensors, y=intensities, marker="o", sort=False, color="firebrick")

# Add title and labels
plt.title('Intensity at each sensor compared to sensor 1', fontsize=18)
plt.xlabel('Sensor', fontsize=15)
plt.ylabel('Intensity (%)', fontsize=15)
plt.savefig("intensity_all.pdf")
# Show the plot
plt.show()
```
