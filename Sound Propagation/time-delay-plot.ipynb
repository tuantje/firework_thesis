```python
import matplotlib.pyplot as plt

# Time delays
delays = [1.0549206349206348, 0.5545124716553288, 0.6453514739229025, 1.1741496598639456, 
          1.9663492063492063, 0.7064852607709751, -734.3767346938776, 510.885]

# Sensor labels
sensors = ['s1-s2', 's1-s3', 's1-s4', 's1-s5', 's1-s6', 's1-s7', 's1-s8', 's1-s9']

plt.figure(figsize=(10, 6))
plt.plot(sensors, delays, color='firebrick')
plt.xlabel('Sensor Pairs')
plt.ylabel('Time Delay (seconds)')
plt.title('Time Delays for Each Sensor Relative to Sensor 1')
plt.tight_layout()
plt.show()
```
