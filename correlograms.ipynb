```python
import numpy as np
import matplotlib.pyplot as plt
import librosa

# Load sensor 1 audio file
y1, sr1 = librosa.load('/home/tnguyen/Downloads/sensor1.wav')

# Take a subset of the data for visualization purposes
y1 = y1[:10000]

# Create a list of the sensor numbers we're interested in
other_sensors = [2, 5, 9]

# Create the 3x1 subplot figure
fig, ax = plt.subplots(3, 1, figsize=(10,15))

for i, sensor in enumerate(other_sensors):
    # Load the audio file for the current sensor
    yi, sri = librosa.load(f'/home/tnguyen/Downloads/sensor{sensor}.wav')

    # Take a subset of the data
    yi = yi[:10000]

    # Compute and plot cross-correlation
    ax[i].xcorr(y1, yi, usevlines=True, maxlags=50, normed=True, lw=2, color='firebrick')
    ax[i].grid(True)
    ax[i].set_title(f'Cross Correlation Sensor 1 and Sensor {sensor}')

# Adjust the layout and display the plot
plt.tight_layout()
# Save the figure
plt.savefig('correlograms.pdf')
plt.show()


```
